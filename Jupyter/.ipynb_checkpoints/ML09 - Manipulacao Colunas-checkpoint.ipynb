```python
#abrindo arquivos no pandas
import pandas as pd

#arquivos csv usar pd.read_csv
#arquivos excel usar pd.read_excel
dados = pd.read_csv('C:/Users/Win7/Documents/Tabelas/athlete_events.csv')
```

```python
#Verificando o arquivo carregado
dados.head()
```

```python
#alterando o nome da coluna
#colocando o comando inplace = True não é exibido o dataset após realizar o comando abaixo
dados.rename(columns={'Name':'Nome','Sex':'Sexo','Age':'Idade'}, inplace = True)
```

```python
#filtra apenas a coluna Height
dados['Height']
```

```python
#Realiza a contagem de cada dado na coluna
dados['City'].value_counts()
```

```python
#Realiza calculos com os dados da planilha carregada
dados.describe()
```

```python
#Excluir colunas de um dataframe
#axis = 0 exclui linha
#axis = 1 exclui Coluna
dados.drop('ID', axis = 1, inplace = True)
dados.drop('Season', axis = 1, inplace = True)
dados.drop('City', axis = 1, inplace = True)
```

```python
#plota o gráfico
#bins = numero de barras

#caso for importar de um array numpy usar o comando plt.hist(nome_array,bins=numero de barras)
dados.hist(column='Idade', bins=80)
plt.show()
```
